# Bar data

```python
from ib_async import *
util.startLoop()

ib = IB()
ib.connect('127.0.0.1', 7497, clientId=14)
```

```python
t=Stock(symbol='TSLA',exchange='SMART',currency='USD')
tsla=ib.qualifyContracts(t)[0]
tsla
```

```python
data=ib.reqHistoricalData(contract=tsla,endDateTime="",durationStr='2 D',barSizeSetting='1 min',whatToShow='MIDPOINT',useRTH=True,formatDate=2)
data
```

```python
data=ib.reqHistoricalData(Stock('SPY','SMART','USD'),"",'10 D','1 min','TRADES',True,1)
data
```
